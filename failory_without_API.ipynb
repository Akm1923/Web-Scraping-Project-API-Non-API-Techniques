```python
city=[]
started_in_year=[]
founders=[]
Industries=[]
employees=[]
funding=[]
funding_rounds=[]
investors=[]
state=[]


l=soup.find_all("li")

for i in l:
  # print(i.text.split(":"))
  string_list=i.text.split(":")
  if(string_list[0]=="State"):
    state.append(string_list[1])
  if(string_list[0]=="City"):
    city.append(string_list[1])
  if(string_list[0]=="Started in"):
    started_in_year.append(string_list[1])
  if(string_list[0]=="Founders"):
    founders.append(string_list[1])
  if(string_list[0]=="Industries"):
    Industries.append(string_list[1])
  if(string_list[0]=="Number of employees"):
    employees.append(string_list[1])
  if(string_list[0]=="Funding"):
    funding.append(string_list[1])
  if(string_list[0]=="Funding rounds"):
    funding_rounds.append(string_list[1])
  if(string_list[0]=="Number of investors"):
    investors.append(string_list[1])
```

Some of the values are not present there so the above code will not give the required result (in founders,employees,investors)

```python
city=[]
started_in_year=[]
founders=[]
Industries=[]
employees=[]
funding=[]
funding_rounds=[]
investors=[]

data=re.split(r'State:', text)[1:]
for i in data:
  founders.append(re.findall(r'Founders:\s*(.*)', i))
  city.append(re.findall(r'City:\s*(.*)', i))
  started_in_year.append(re.findall(r'Started in:\s*(.*)', i))
  Industries.append(re.findall(r'Industries:\s*(.*)', i))
  employees.append(re.findall(r'Number of employees:\s*(.*)', i))
  funding.append(re.findall(r'Funding:\s*(.*)', i))
  funding_rounds.append(re.findall(r'Funding rounds:\s*(.*)', i))
  investors.append(re.findall(r'Number of investors:\s*(.*)', i))
```

The above code splits the text along with word "state" and appends the value of particular company at a particular index and if it not present then filling it with [ ] which we will further replace by nan

```python
len(city),len(started_in_year),len(founders),len(Industries),len(employees),len(funding),len(funding_rounds),len(investors)
```

As [ ] is present now, replacing it with nan


```python
lists = [city, started_in_year, founders, Industries, employees, funding, funding_rounds, investors]
for i in range(len(lists)):
    for j in range(len(lists[i])):
        if lists[i][j] == []:
            lists[i][j] = np.nan
```

```python
founders[90:120]
```

```python
founders = [', '.join(f) if isinstance(f, list) else np.nan for f in founders]
employees= [', '.join(f) if isinstance(f, list) else np.nan for f in employees]
investors=[', '.join(f) if isinstance(f, list) else np.nan for f in investors]
city=[', '.join(f) if isinstance(f, list) else np.nan for f in city]
started_in_year=[', '.join(f) if isinstance(f, list) else np.nan for f in started_in_year]
Industries=[', '.join(f) if isinstance(f, list) else np.nan for f in Industries]
funding=[', '.join(f) if isinstance(f, list) else np.nan for f in funding]
funding_rounds=[', '.join(f) if isinstance(f, list) else np.nan for f in funding_rounds]
```

For Company name:

```python
company_names=[]
for i in soup.find_all("h3"):
  text=i.text.strip()
  text_new=re.sub("[\d{3}|\d{2}|\d{1})]","",text)
  company_names.append(text_new)
```

```python
company_names[:5],len(company_names)
```

For Company website:

```python
website = [link['href'] for link in soup.find_all('a', target="_blank")]
website=website[1:-4]
```

```python
from tokenize import Ignore
data_dict={"Company":company_names,"State":state,"City":city,"Started in":started_in_year,"Founders":founders,"Industries":Industries,"Number of employees":employees,"Funding":funding,"Funding rounds":funding_rounds,"Number of investors":investors,"Website":website}
df=pd.DataFrame(data_dict)
```
